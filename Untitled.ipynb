```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
from PIL import Image
import matplotlib.style
import matplotlib as mpl
mpl.style.use('classic')



# Membaca citra asli
RGB = cv2.imread('beo.png');
plt.imshow(RGB)
plt.show()
```

```python
# % Melakukan transformasi ruang warna citra yang semula RGB menjadi HSV
HSV = rgb2hsv(RGB);
figure, imshow(HSV);
```

```python
# % Melakukan segmentasi terhadap warna biru (nilai Hue 100-250)
H = HSV(:,:,1);
range = [100 250]/360; 
mask = (H>range(1)) & (H<range(2));
figure, imshow(mask)
```

```python
# % 1. Mengubah warna background menjadi grayscale
H = HSV(:,:,1);
S = HSV(:,:,2);
V = HSV(:,:,3);
 
S = S.*mask;
HSV1 = cat(3,H,S,V);
 
RGB1 = hsv2rgb(HSV1);
figure, imshow(RGB1);
```

```python
% 2. Mengubah warna objek menjadi merah
H = HSV(:,:,1);
S = HSV(:,:,2);
V = HSV(:,:,3);
 
H(mask) = H(mask)+.4;
HSV2 = cat(3,H,S,V);
 
RGB2 = hsv2rgb(HSV2);
figure, imshow(RGB2)
```

```python
% 3. Mengubah warna objek menjadi merah & warna background menjadi grayscale
H = HSV(:,:,1);
S = HSV(:,:,2);
V = HSV(:,:,3);
 
H(mask) = H(mask)+.4;
S = S.*mask;
HSV3 = cat(3,H,S,V);
 
RGB3 = hsv2rgb(HSV3);
figure, imshow(RGB3);
```
